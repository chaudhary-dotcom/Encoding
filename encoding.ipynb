# Using Ordinal Encoding first
It is used in input feature having categories in order format

```python
df.head()
```

According to dataset we have 4 categorical column. First we have to find the ordinal variable. Which are review, education and purchased. Gender is nominal category. For this thing in review and education we used ordianl encoding and in purchased we used label encoding because label encoding always used in target variable.

###  Using ordinal encoding 

```python
df = df.iloc[:,2:]
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,0:2], df.iloc[:,-1], test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=[['Poor', 'Average', 'Good'], ['School', 'UG', 'PG']])
```

```python
oe.fit(X_train)
```

```python
X_train = oe.transform(X_train)
X_test = oe.transform(X_test)
```

```python
X_train
```

```python
oe.categories_
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
le.fit(y_train)
```

```python
y_train = le.transform(y_train)
y_test = le.transform(y_test)
```

## 3. OneHotEncoding Using Sklearn

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_car.iloc[:, 0:4], df_car.iloc[:,-1], test_size=0.2, random_state=2)
```

```python
X_train.shape
```

```python
ohe = OneHotEncoder(drop='first', sparse_output=False, dtype=int)
```

```python
X_train_new = ohe.fit_transform(X_train[['fuel', 'owner']])
X_test_new = ohe.transform(X_test[['fuel', 'owner']])
```

```python
# np.hastack() stands for horizontal stack in Numpy. It concatenates arrays side by side (column-wise)
combined = np.hstack([X_train[['brand', 'km_driven']].values, X_train_new])
```

### 4. OneHotEncoding with Top Categories

# Applying Column Transformer

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(transformers=[
    ('tnf1', OneHotEncoder(sparse_output=False, drop='first',handle_unknown='ignore'), ['brand', 'fuel', 'owner'])
], remainder='passthrough'

)
```

```python
transformer.fit_transform(X_train).shape
```
